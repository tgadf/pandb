# Genius Match

```python
artistNameData  = mu.getArtistNameData("Genius", "Search")
numAlbumsData   = mu.getNumAlbumsData("Genius", "Search")

artistNameData.name = "ArtistName"
numAlbumsData.name  = "NumAlbums"
artistBasicData = DataFrame(artistNameData).join(numAlbumsData).sort_values(by="NumAlbums", ascending=False)
```

```python
meu = masterManualEntriesUtils()
mmeDF = meu.getDF()
knownIDs  = mmeDF["Genius"][mmeDF["Genius"].notna()]
unknownDF = mmeDF[mmeDF["Genius"].isna()]
print(artistBasicData.shape)
unmatched = artistBasicData[~artistBasicData.index.isin(knownIDs)]

geniusAlbums = mu.getArtistAlbumsData('Genius', 'DB')
```

```python
from matchArtistToDB import masterMatch
```

```python
mm = masterMatch()
unMatchedMatches = {}
```

```python
ts = timestat("Matching Genius")
N = 25000
for i,(idx,row) in enumerate(unmatched.iterrows()):
    if i < 2500:
        continue
    if i >= N:
        break
    geniusName = row["ArtistName"]
    results = mm.matchNames(tomatch=unknownDF["ArtistName"], value=geniusName)    
    possible = results[results > 0.9].sort_values(ascending=False)
    if len(possible) > 0:
        unMatchedMatches[idx] = possible
    if (i+1) % 250 == 0 or (i+1) == 50:
        ts.update(n=i+1,N=N)
ts.stop()
```

```python
unMatchedResults = DataFrame(Series(unMatchedMatches, name="Matches"))
unMatchedResults["NumMatches"] = unMatchedResults["Matches"].apply(len)
unMatchedResults["Max"] = unMatchedResults["Matches"].apply(max)
unMatchedResults = unMatchedResults.sort_values(by="Max", ascending=False)
unMatchedResults = unMatchedResults.join(unmatched)
```

```python
import regex

def getLetter(value):
    return regex.compile(r"[^\w\d\s:\p{L}]+").search(value)
def getAscii(value):
    return regex.compile(r"[[:ascii:]]+").search(value)
def getLatin(value):
    return regex.compile(r"[\p{IsLatin}]+").search(value)
def getKorean(value):
    return regex.compile(r"[\p{IsHangul}]+").search(value)
def getThai(value):
    return regex.compile(r"[\p{IsThai}]+").search(value)
def getHebrew(value):
    return regex.compile(r"[\p{IsHebrew}]+").search(value)
def getRussian(value):
    return regex.compile(r"[\p{IsCyrillic}]+").search(value)
def getGreek(value):
    return regex.compile(r"[α-ωΑ-Ω]+").search(value)
def getArabic(value):
    return regex.compile(r"^[\u0621-\u064A\u0660-\u0669]+$").search(value)
def getArmenian(value):
    return regex.compile(r"^[\u0561-\u0587\u0531-\u0556]+$").search(value)
def getChinese(value):
    return regex.compile(r"[\p{IsHan}]+").search(value)
def getJapanese(value):
    return regex.compile(r"[\p{IsHira}\p{IsKatakana}]+").search(value)
def getJCese(value):
    return regex.compile(r"[\p{IsHan}\p{IsBopo}\p{IsHira}\p{IsKatakana}]+").search(value)    
def getVietnamese(value):
    return regex.compile(r"[ườầ]+").search(value)
def getSlavic(value):
    return regex.compile(r"[ĆćĚěŠšČčŘřŽžÝýďŤťĎŇňŮů]+").search(value)
def getGerman(value):
    return regex.compile(r"[ß]+").search(value)
def getScandinavian(value):
    return regex.compile(r"[æøåÆØÅ]+").search(value)
def getIcelandic(value):
    return regex.compile(r"[ðÐþÞ]+").search(value)
def getPolish(value):
    return regex.compile(r"[ĄąĘęŁłŃńŚśŹźŻż]+").search(value)
def getFrench(value):
    return regex.compile(r"[ÂâÊêËëÎîÏïÔôÛûŸÿçœ]+").search(value)
def getPortuguese(value):
    return regex.compile(r"[ÃãÕõ]+").search(value)
def getSpanish(value):
    return regex.compile(r"[¡ñ]+").search(value)
def getSpecial(value):
    return regex.compile(r"[‒’„“‐ʻēū]+").search(value)
def getRomanian(value):
    return regex.compile(r"[ĂăȘșȚț]+").search(value)
def getHungarian(value):
    return regex.compile(r"[Őő]+").search(value)
def getUmlaut(value):
    return regex.compile(r"[äöüÄÖÜ]+").search(value)
def getAccentAcute(value):
    return regex.compile(r"[ÁáÉéÍíÓóÚú]+").search(value)
def getAccentGrave(value):
    return regex.compile(r"[ÀàÈèÌìÒòÙù]+").search(value)
    
def getDJMC(value):
    return sum([value.startswith(y) for y in ["DJ ", "Dj ", " DJ", " Dj", "Lil ", "Lil' ", "MC ", "D.J."]]) > 0
def getClassical(value):
    return sum([value.find(y) != -1 for y in ["Theatre", "Theater", "Chorus", "Orchestra"]]) > 0
def getAnd(value):
    return sum([value.find(y) != -1 for y in [" And ", " and ", " & "]]) > 0
def getSlash(value):
    return sum([value.find(y) != -1 for y in ["+", "/", "*", "?", "'"]]) > 0
def getDash(value):
    return sum([value.find(y) != -1 for y in ["-", ".", ","]]) > 0
def getNumerical(value):
    return value.isdigit()
def getSingleWord(value):
    return len(value.split()) == 1
def getFourWords(value):
    return len(value.split()) >= 4
    



def getNum(x):
    if getKorean(x) is not None:
        return "Korean"
    elif getJapanese(x) is not None:
        return "Japanese"
    elif getChinese(x) is not None:
        return "Chinese"
    elif getJCese(x) is not None:
        return "Xese"
    elif getThai(x) is not None:
        return "Thai"
    elif getVietnamese(x) is not None:
        return "Vietnamese"
    elif getHebrew(x) is not None:
        return "Hebrew"
    elif getArabic(x) is not None:
        return "Arabic"
    elif getArmenian(x) is not None:
        return "Armenian"
    elif getRussian(x) is not None:
        return "Russian"
    elif getGreek(x) is not None:
        return "Greek"
    elif getIcelandic(x) is not None:
        return "Icelandic"
    elif getRomanian(x) is not None:
        return "Romanian"
    elif getHungarian(x) is not None:
        return "Hungarian"
    elif getScandinavian(x) is not None:
        return "Scandinavian"
    elif getPolish(x) is not None:
        return "Polish"
    elif getSlavic(x) is not None:
        return "Slavic"
    elif getUmlaut(x) is not None:
        return "Umlaut"
    elif getGerman(x) is not None:
        return "German"
    elif getSpanish(x) is not None:
        return "Spanish"
    elif getPortuguese(x) is not None:
        return "Portuguese"
    elif getAccentGrave(x) is not None:
        return "Grave"
    elif getAccentAcute(x) is not None:
        return "Acute"
    elif getSpecial(x) is not None:
        return "Special"
    else:
        if getDJMC(x):
            return "DJMC"
        elif getClassical(x):
            return "Classical"
        elif getAnd(x):
            return "And"
        elif getSlash(x):
            return "Slash"
        elif getDash(x):
            return "Dash"
        elif getNumerical(x):
            return "Numerical"
        elif getSingleWord(x):
            return "SingleWord"
        elif getFourWords(x):
            return "FourWords"
        elif getLatin(x) is not None:
            return "Latin"
        else:
            return "Regular"
```

```python
from time import sleep
from googletrans import Translator
translator = Translator()

maxN = 1000
namesToDo = Series([k for k in namesToTranslate.keys() if translatedNames.get(k) is None]).head(maxN)
ts = timestat("Translating {0}/{1} names.".format(len(namesToDo),len(namesToTranslate)))
errs = 0

namesToDoTranslated = {}
for i,dbName in enumerate(namesToDo):
    print(i,'\t',dbName)
    try:
        result = translator.translate(dbName, dest='en').text
    except:
        print("Error translating [{0}] [{1}]".format(i,dbName))
        errs += 1
        sleep(5)
        
    if errs >= 3:
        print("Stopping after {0} errors".format(errs))
        break
        
    namesToDoTranslated[dbName] = result
    sleep(3)
    
    if (i+1) % 5 == 0:
        ts.update(n=i+1, N=len(namesToDo))
        
ts.stop()
```
