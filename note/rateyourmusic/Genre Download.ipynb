```python
def getNotGenre(x):
    if isinstance(x, list):
        for genre in x:
            if genre.find("Rock") != -1:
                return True
            if genre.find("Pop") != -1:
                return True
    return False

def getGenre(x, genreList=[]):
    if isinstance(x, list):
        retval = sum([genre in x for genre in genreList]) > 0
        return retval
    return False
```

# Compute Relationships

```python
basicData   = DataFrame(mdbio.data.getSummaryNameData()).join(mdbio.data.getSummaryCountsData())
membersData = mdbio.data.getSummaryLinkData()
names = basicData["Name"].to_dict()
asaData = membersData[membersData["Members"].notna()]["AlsoKnownAs"]
memData = membersData[membersData["Members"].notna()]["Members"]
mofData = membersData[membersData["MemberOf"].notna()]["MemberOf"]
rarData = membersData[membersData["RelatedArtists"].notna()]["RelatedArtists"]
```

```python
def getAsa(rData):
    return rData.profile.general['Also Known As']
def getMem(rData):
    return rData.profile.general['Members']
def getMof(rData):
    return rData.profile.extra['Member of']
def getRar(rData):
    return rData.profile.extra['Related Artists']

def getID(x):
    retval = None
    if isinstance(x, list):
        y = {}
        for item in x:
            if isinstance(item, dict):
                y.update({name: dbid for name, dbid in item.items()})
        x = y if len(y) > 0 else None
    if isinstance(x, dict):
        kvals  = {(name,dbid): names.get(dbid) for name,dbid in x.items() if dbid is not None}
        known  = {k: v for k,v in kvals.items() if v is not None}
        unkwn  = [k for k,v in kvals.items() if v is None]
        retval = {name: dbid for (name,dbid) in unkwn} if (len(known) == 0 and len(unkwn) > 0) else None
    return retval

toget = asaData.apply(getID)
asaToGet = basicData.join(toget[toget.notna()], how='right').sort_values(by=["Album","SingleEP"], ascending=False)
asaToGetRefs = Series(dtype='object')

toget = memData.apply(getID)
memToGet = basicData.join(toget[toget.notna()], how='right').sort_values(by=["Album","SingleEP"], ascending=False)
memToGetRefs = Series(dtype='object')

toget = mofData.apply(getID)
mofToGet = basicData.join(toget[toget.notna()], how='right').sort_values(by=["Album","SingleEP"], ascending=False)
mofToGetRefs = Series(dtype='object')

toget = rarData.apply(getID)
rarToGet = basicData.join(toget[toget.notna()], how='right').sort_values(by=["Album","SingleEP"], ascending=False)
rarToGetRefs = Series(dtype='object')
```

```python
print("{0: <10}{1: <10}{2: <10}{3: <10}{4: <10}".format("Mod","Asa", "Mem", "Mof", "Rar"))

for modVal in range(100):
    modValData = mdbio.data.getModValData(modVal)
    
    idx = asaToGet[asaToGet.index.map(mdbio.getModVal) == modVal].index
    asaToGetRefs = concat([asaToGetRefs, modValData.loc[idx].apply(getAsa)])
    
    idx = memToGet[memToGet.index.map(mdbio.getModVal) == modVal].index
    memToGetRefs = concat([memToGetRefs, modValData.loc[idx].apply(getMem)])
    
    idx = mofToGet[mofToGet.index.map(mdbio.getModVal) == modVal].index
    mofToGetRefs = concat([mofToGetRefs, modValData.loc[idx].apply(getMof)])
    
    idx = rarToGet[rarToGet.index.map(mdbio.getModVal) == modVal].index
    rarToGetRefs = concat([rarToGetRefs, modValData.loc[idx].apply(getRar)])
    
    if (modVal+1) % 10 == 0:
        print("{0: <10}{1: <10}{2: <10}{3: <10}{4: <10}".format(modVal+1,len(asaToGetRefs), len(memToGetRefs), len(mofToGetRefs), len(rarToGetRefs)))
    
permDir = MusicDBDir(permDBDir)
asaToGetRefs.name = "Refs"
asaToGet = asaToGet.join(asaToGetRefs)
asaToGet["Metric"] = asaToGet["Album"]*asaToGet["SingleEP"]
mdbd = MusicDBData(path=permDir, fname="asaToGet")
mdbd.save(data=asaToGet)

memToGetRefs.name = "Refs"
memToGet = memToGet.join(memToGetRefs)
memToGet["Metric"] = memToGet["Album"]*memToGet["SingleEP"]
mdbd = MusicDBData(path=permDir, fname="memToGet")
mdbd.save(data=memToGet)

mofToGetRefs.name = "Refs"
mofToGet = mofToGet.join(mofToGetRefs)
mofToGet["Metric"] = mofToGet["Album"]*mofToGet["SingleEP"]
mdbd = MusicDBData(path=permDir, fname="mofToGet")
mdbd.save(data=mofToGet)

rarToGetRefs.name = "Refs"
rarToGet = rarToGet.join(rarToGetRefs)
rarToGet["Metric"] = rarToGet["Album"]*rarToGet["SingleEP"]
mdbd = MusicDBData(path=permDir, fname="rarToGet")
mdbd.save(data=rarToGet)
```
